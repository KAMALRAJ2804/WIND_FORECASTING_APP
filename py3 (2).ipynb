```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Sample structure (replace this with your real dataset)
# Example: df_combined = pd.read_csv("your_cleaned_combined_dataset.csv")

# Simulated dataset for illustration
np.random.seed(42)
df_combined = pd.DataFrame({
    'Pressure': np.random.uniform(990, 1020, 1000),
    'Temp_5m': np.random.uniform(20, 35, 1000),
    '20m_Avg': np.random.uniform(2, 10, 1000),
    '100m_N_Avg': np.random.uniform(4, 15, 1000)
})

# --- Feature Selection ---
features = ['Pressure', 'Temp_5m', '20m_Avg']
target = '100m_N_Avg'

# --- Scaling ---
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df_combined[features + [target]])
X_scaled = scaled[:, :-1]
y_scaled = scaled[:, -1]

# --- Create Sequences ---
def create_lstm_sequences(X, y, time_steps=10):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:i+time_steps])
        ys.append(y[i+time_steps])
    return np.array(Xs), np.array(ys)

X_seq, y_seq = create_lstm_sequences(X_scaled, y_scaled, time_steps=10)

# --- Train-Test Split ---
X_train, X_test, y_train, y_test = train_test_split(X_seq, y_seq, test_size=0.2, random_state=42)

# --- Build LSTM Model ---
model = Sequential()
model.add(LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# --- Train the Model ---
model.fit(X_train, y_train, epochs=20, batch_size=16, validation_split=0.2, verbose=1)

# --- Predictions ---
y_pred_scaled = model.predict(X_test)

# Combine X and y for inverse scaling
X_test_last = X_test[:, -1]
y_pred_combined = np.hstack((X_test_last, y_pred_scaled))
y_test_combined = np.hstack((X_test_last, y_test.reshape(-1, 1)))

# Recover original values
y_pred_final = scaler.inverse_transform(y_pred_combined)[:, -1]
y_true_final = scaler.inverse_transform(y_test_combined)[:, -1]

# --- Evaluation ---
mae_lstm = mean_absolute_error(y_true_final, y_pred_final)
r2_lstm = r2_score(y_true_final, y_pred_final)


print("✅ LSTM Results:")
print("MAE:", round(mae_lstm, 3))

print("R² Score:", round(r2_lstm, 3))
```

```python
%%writefile score.py
import joblib
import pandas as pd
import json
import numpy as np
from azureml.core.model import Model

def init():
    global model
    # Load the model from the registered path
    model_path = Model.get_model_path('xgb_wind_model')
    model = joblib.load(model_path)

def run(raw_data):
    try:
        data = json.loads(raw_data)
        df = pd.DataFrame(data)
        prediction = model.predict(df)
        return prediction.tolist()
    except Exception as e:
        return str(e)
```

```python
%%writefile app.py
import streamlit as st
import joblib
import numpy as np

# Load model
model = joblib.load('xgb_wind_model.pkl')

# Title
st.title("🌬️ Wind Speed Forecasting using XGBoost")

# Input fields
pressure = st.number_input("Pressure (mbar)", min_value=900.0, max_value=1100.0, value=1010.0)
temperature = st.number_input("Temperature (°C)", min_value=0.0, max_value=50.0, value=25.0)
wind_20m = st.number_input("20m Wind Speed (m/s)", min_value=0.0, max_value=25.0, value=5.0)

# Predict button
if st.button("Predict 100m_N_Avg Wind Speed"):
    input_data = np.array([[pressure, temperature, wind_20m]])
    prediction = model.predict(input_data)[0]
    st.success(f"💨 Predicted Wind Speed at 100m_N: {round(prediction, 2)} m/s")
```
